### 문제1. 결측치가 있는 데이터의 행(레코드)을 제거하고, 앞에서부터 70% 데이터만 활용해 'f1' 컬럼 1사분위 값을 구하시오. (소수점 제거 후 정수 출력)

```python
import pandas as pd
df = pd.read_csv("C:/Users/WJM/Downloads/members.csv")

# print(df)
df = df.dropna()

df = df.iloc[:int(len(df)*0.7)]

result = df['f1'].quantile(.25)

print(result)
```

### 문제2. 2000년 데이터 중 2000년 평균보다 큰 값의 데이터 수를 구하시오. (index: 연도)

```python
import pandas as pd
df = pd.read_csv('C:/Users/WJM/Downloads/year.csv', index_col='Unnamed: 0')

# print(df)
# print(df.info())
# print(df.describe())

m = df.loc[2000].mean()
# print(m)

result = sum(df.loc[2000, :] > m)

print(result)
```

# 제5회 기출 문제

## 작업형 제2유형

### 문제. 차량의 특징 기반 데이터를 활용해 test 데이터의 중고차 가격을 예측하시오.
#### 학습용 데이터(train.csv)를 이용하여 차량 가격을 예측하는 모델을 만든 후 이를 평가용 데이터(test.csv)에 적용해 얻은 예측값을 다음과 같은 형식의 CSV 파일로 생성하시오.
#### 제출 파일은 pred(예측된 차량 가격) 컬럼을 포함해 result.csv라는 이름으로 제출해야 한다.
#### 제출한 모델의 성능은 RMSE(Root Mean Square Error) 평가지표에 따라 채점한다.

```python
import pandas as pd
train = pd.read_csv(r"C:\Users\WJM\Downloads\5회\train.csv")
test = pd.read_csv(r"C:\Users\WJM\Downloads\5회\test.csv")

# 탐색적 데이터 분석
# print(train.head())
# print("\n")
# print(train.info())
# print("\n")
# print(train.describe())
# print(test.info())

# 데이터 전처리
# 훈련용 데이터셋에서 종속변수 분리
target = train.pop('price')
train = pd.get_dummies(train)
test = pd.get_dummies(test)

# 데이터 분할
from sklearn.model_selection import train_test_split
X_tr, X_val, y_tr, y_val = train_test_split(train, target, test_size=0.2, random_state=123)

# 머신러닝 모델링, 학습 및 평가
from sklearn.metrics import root_mean_squared_error
from xgboost import XGBRegressor
import xgboost as xgb
xg = xgb.XGBRegressor(random_state=123)
xg.fit(X_tr, y_tr)
pred = xg.predict(X_val)
root_mean_squared_error(y_val, pred)

pred = xg.predict(test)
result = pd.DataFrame({'pred':pred})
result.to_csv("result.csv", index = False)

pd.read_csv("result.csv")
```
